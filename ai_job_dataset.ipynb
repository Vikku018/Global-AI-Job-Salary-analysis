```python
import pandas as pd
```

```python
df=pd.read_csv("ai_job_dataset.csv")
```

```python
df["experience_level"].unique()
```

```python
df["experience_level"]=df["experience_level"].replace({'SE':"Senior",
                                'EN':"Entry",
                                'MI':"Mid",
                                'EX':"Executive"
                                })
```

```python
df["employment_type"].unique()
```

```python
df["employment_type"]=df["employment_type"].replace({"CT":"Contract",
                               "FL":"Freelance",
                               "PT":"Part-time",
                               "FT":"Full-time"
                               })
```

```python
df["company_size"].unique()
```

```python
df["company_size"]=df["company_size"].replace({"M":"Medium",
                            "L":"Large",
                            "S":"Small"})
```

```python
df = df.rename(columns={"remote_ratio": "work_mode",
                        "salary_usd":"salary"
                        })
```

```python
df["work_mode"].unique()
```

```python
df["work_mode"]=df["work_mode"].replace({0:"On-Site",
                        50:"Hybrid",
                         100:"Remote"})
```

```python
df.head(1)
```

```python
df["required_skills"].head()
```

```python
# df[["skill_1", "skill_2","skill_3","skill_4","skill_5"]] = df["required_skills"].str.split(",",expand=True)
```

```python
df["required_skills"] = df["required_skills"].str.split(",")
```

```python
df=df.explode("required_skills")
```

```python
df["required_skills"] = df["required_skills"].str.strip()
```

```python
df = df.dropna(subset=["required_skills"])
```

```python
skill_count = df["required_skills"].value_counts().reset_index()
skill_count.columns = ["skill", "frequency"]
print(skill_count.head())
```

```python
df.columns
```

```python
df=df.drop(['salary_currency','years_experience', 'industry', 'posting_date',
       'application_deadline', 'job_description_length', 'benefits_score'], axis=1)
```
